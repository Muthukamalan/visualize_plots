# Pie chart

```python
df = pd.read_csv("./data/pie_chart_data.csv")
df.sample(3)
```

```python
plt.pie(
    x=df['Number of Cars'],
    labels=df['Engine Fuel Type'].values,
    autopct="%.2f%%", # 2 decimal
    textprops = {
        "size":"large",
        "fontweight":"20",
        'rotation':'horizontal',
        'color':'w'
    }
)
plt.legend(loc="best");
plt.title("Cars by Engine Fuel Type")
plt.show()
```

# Stacked Area chart plot

```python
df = pd.read_csv("./data/stacked_area_chart_data.csv")
df.sample(3)
```

```python
# remove border
sns.despine(left=True)
```

```python
plt.figure(figsize=(15,4))
labels = df.drop(['Year','Other'],axis=1).columns
colors = ['#011638','#7e2987','#ef2026']
plt.stackplot(df['Year'], # X-axis

    # stacking parameters
    df['Gas'], 
    df['Petrol'],
    df['Diesel'],
    colors=colors,
    edgecolor="none",
    labels=labels
)

plt.ylabel("Number of Cars",fontsize=10)
plt.title("Popularity of Engine Fuel Type (1982-2016)",weight="bold",fontsize=12)
plt.xticks(df['Year'],rotation=90)
plt.legend()
plt.show()
```

# Line Plot

```python
df = pd.read_csv("./data/line_chart_data.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.head(3)
```

# Scatter plot

```python
df = pd.read_csv('./data/scatter_data.csv')
df.sample(3)
```

```python
plt.figure(figsize=(9,6))
scatter = plt.scatter(
    x=df['Area (ft.)'],
    y=df['Price'],
    alpha=0.6,                  # alpha = transperency
    c =df['Building Type'], # color
    s = df['Property #']  # size
)
plt.legend(*scatter.legend_elements(),title="Building Type")
plt.title("Relation between Area and Price of Califorina Real Estate")
plt.xlabel("Area (ft.)")
plt.ylabel("Price ($)")
plt.show()
```

# Spyder Chart

```python
from math import pi
```

```python
df = pd.DataFrame({
'group': ['A','B','C','D'],
'var1': [38, 1.5, 30, 4],
'var2': [29, 10, 9, 34],
'var3': [8, 39, 23, 24],
'var4': [7, 31, 33, 14],
'var5': [28, 15, 32, 14]
})

df
```

```python
categories=list(df)[1:]
N = len(categories)
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]


print(f"categories:{categories} and N={N} and \nvalues={values}  and \nangles={angles}")
```

```python
ax = plt.subplot(111, polar=True) 
plt.xticks(angles[:-1], categories, color='grey', size=8) 
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)
ax.plot(angles, values, linewidth=1, linestyle='solid')
ax.fill(angles, values, 'b', alpha=0.1)  
plt.show()
```
