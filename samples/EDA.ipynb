```python
from IPython.display import  Image,display,Markdown
import os
import warnings
```

```python
images = os.listdir('./images/')
total_images = len(images)
counter= 1
for i in images:
    display(Markdown(f"<center><b align='bold'>{i}</b></center>"))
    display(Image(filename=f"./images/{i}",alt=i))
    counter+=1
    print("\n\n\n")
    
    
if counter==total_images:
    warnings.warn("Some images are not displayed properly")
```
